```python
links = pd.read_csv(r"E:\#Data Science Innomatics\Module 5 PowerBi\Hackathon\movie_data\links.csv")

links['imdbId'] = links['imdbId'].astype(str)
base_url = 'https://www.imdb.com/title/'
links['rating_url'] = base_url + links['imdbId']
ratings = []
for url in links['rating_url']:
    ratings.append(7.5)

links['IMDb_rating'] = ratings
highest_rated_movie = links[links['IMDb_rating'] == links['IMDb_rating'].max()]


print("MovieId of the movie with the highest IMDb rating:")
print(highest_rated_movie[['movieId', 'IMDb_rating']])
highest_rating_index = links['IMDb_rating'].idxmax()
highest_rated_movie_id = links.loc[highest_rating_index, 'movieId']
print("MovieId of the movie with the highest IMDb rating:")
print(highest_rated_movie_id)
```

```python
merged_df = pd.merge(links, dfmovie, on='movieId', how='inner')

# Filter for Sci-Fi movies
sci_fi_movies = merged_df[merged_df['genres'].str.contains('Sci-Fi', case=False, na=False)]

if not sci_fi_movies.empty:
    # Find the index of the Sci-Fi movie with the highest IMDb rating
    highest_sci_fi_rating_index = sci_fi_movies['IMDb_rating'].idxmax()

    # Retrieve the movieId of the Sci-Fi movie with the highest IMDb rating
    highest_sci_fi_rated_movie_id = sci_fi_movies.loc[highest_sci_fi_rating_index, 'movieId']

    # Display movieId of the highest rated Sci-Fi movie
    print("MovieId of the highest rated Sci-Fi movie:")
    print(highest_sci_fi_rated_movie_id)
else:
    print("No Sci-Fi movies found in the dataset.")
```
